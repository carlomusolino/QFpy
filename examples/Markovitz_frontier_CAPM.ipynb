```python
import numpy as np
import matplotlib.pyplot as plt

from QFpy.capm_utils import find_efficient_frontier, compute_volatility

# Set the default font family and size
plt.rcParams['font.family'] = 'serif'  # Example: 'serif', 'sans-serif', 'monospace', etc.
plt.rcParams['text.usetex'] = True
plt.rcParams['font.serif'] = ['Computer Modern']  # Specify the font family if using 'serif'
plt.rcParams['font.size'] = 14  # Set the font size
plt.rcParams['xtick.labelsize'] = 14  # Replace 14 with the desired font size
```

```python
rewards = np.linspace(0.,0.5,100)
sigma = np.array([0.2,0.3,0.25,0.35])
mu = np.array([0.1,0.15,0.18,0.2])
rho = np.array([[1,0.2,0.3,0.1],
                [0.2,1.,0.05,0.1],
                [0.3,0.05,1,0.05],
                [0.1,0.1,0.05,1]])

print(f"Correlations:\n\n{rho}\n")
print(f"Vols:\n\n{sigma}\n")

    
weights = [
    find_efficient_frontier(r,sigma,mu,rho) for r in rewards
]
risk = [ compute_volatility(w, sigma, rho) for w in weights ]
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)

ax.scatter(sigma,mu)
ax.plot(risk,rewards,'-', color="black", lw=2)

ax.text(0.32,0.36, "Efficient frontier")

ax.set_xlabel("Risk")
ax.set_ylabel("Return")
```
